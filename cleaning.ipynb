```python
import pandas as pd
```

```python
df = pd.read_csv('AB_NYC_2019.csv')
```

```python
df.head() 
```

```python
df.isna().sum()
```

```python
df['last_review'] = pd.to_datetime(df['last_review'])
```

```python
df[df['last_review'].isna()]
```

```python
last_date = df['last_review'].max()
```

# **Adding new columns**

```python
df['days_since_last_review'] = (last_date - df['last_review']).dt.days
```

```python
df['has_reviews'] = df['last_review'].notna()
```

```python
df[(df['name'].isna() & df['last_review'].isna())]
```

# **removing rows if name and last_review both are null**

```python
df = df[~(df['name'].isna() & df['last_review'].isna())]        # ~ means NOT
```

```python
df.info()
```

```python
df['name'] = df['name'].str.strip()
df['host_name'] = df['host_name'].str.strip()
```

```python
df[['price', 'number_of_reviews', 'price', 'calculated_host_listings_count', 'availability_365']].describe()
```

```python
df['neighbourhood'].value_counts()
```

# **Finding and Creating superhosts column**

```python
df['superhost'] = df['calculated_host_listings_count'].apply(lambda x: True if x> df['calculated_host_listings_count'].quantile(0.96) else False)
```

# Cleaning the has_reviews column

```python
df['has_reviews'] = df['has_reviews'].apply(lambda x: 'has reviews' if x==True else 'No reviews')
```
